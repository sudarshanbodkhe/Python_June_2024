```python
def add(n1,n2):
    add=n1+n2
    print(f"The addition is:{add}")
def sub(n1,n2):
    sub=n1-n2
    print(f"The substraction is:{sub}")
def mul(n1,n2):
    mul=n1*n2
    print(f"The multiplication is:{mul}")
def div(n1,n2):
    div=n1/n2
    print(f"The division is:{div}")

def calculator():
    print("Enter option 1 for Addition")
    print("Enter option 2 for Substraction")
    print("Enter option 3 for Multiplication")
    print("Enter option 4 for Division")
    option=eval(input("Enter the option(1-4):"))
    n1=eval(input("Enter number1:"))
    n2=eval(input("Enter number2:"))

    if option==1:
        add(n1,n2)
    elif option==2:
        sub(n1,n2)
    elif option==3:
        mul(n1,n2)
    elif option==4:
        div(n1,n2)
    else:
        print("Invalid input")

    
calculator()
    
    
```
